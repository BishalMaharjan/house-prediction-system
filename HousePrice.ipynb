```python
house_area = []
for a in range(df.shape[0]):
  b = df.iloc[a].area_covered

  if df.iloc[a].area_units == 'Aana':
    if re.match(r'\b\d+[-.]\d+[-.]\d+[-.]\d+\b', b):
      ropani = float(re.findall(r'\b(\d+)[-.]\d+[-.]\d+[-.]\d+\b', b)[0])
      aana = float(re.findall(r'\b\d+[-.](\d+)[-.]\d+[-.]\d+\b', b)[0])
      paisa = float(re.findall(r'\b\d+[-.]\d+[-.](\d+)[-.]\d+\b', b)[0])
      dam = float(re.findall(r'\b\d+[-.]\d+[-.]\d+[-.](\d+)\b', b)[0])
      area = ropani * 508.737047 + aana * 31.79606544 + paisa * 7.94901636 + dam * 1.98725409
    elif re.match(r'\b\d+[-.]\d+[-.]\d+\b', b):
      aana  = float(re.findall(r'\b(\d+)[-.]\d+[-.]\d+\b', b)[0])
      paisa = float(re.findall(r'\b\d+[-.](\d+)[-.]\d+\b', b)[0])
      dam = float(re.findall(r'\b\d+[-.]\d+[-.](\d+)\b', b)[0])
      area = aana * 31.79606544 + paisa * 7.94901636 + dam * 1.98725409
    elif re.match(r'\b\d+ *[Aa]ana \d+ [Pp]aisa+',b):
      aana = float(re.findall(r'\b(\d+) *[Aa]ana \d+ [Pp]aisa+',b)[0])
      paisa = float(re.findall(r'\b\d+ *[Aa]ana (\d+) [Pp]aisa+',b)[0])
      area = aana * 31.79606544 + paisa * 7.94901636
    elif re.match(r'\b\d+[-.]\d+\b', b):
      aana  = float(re.findall(r'\b(\d+)[-.]\d+\b', b)[0])
      paisa = float(re.findall(r'\b\d+[-.](\d+)\b', b)[0])
      area = aana * 31.79606544 + paisa * 7.94901636 + dam * 1.98725409
    else:
      area = float(b) * 31.79606544

  if df.iloc[a].area_units == 'Ropani':
    if re.match(r'\b\d+-\d+-\d+-\d+\b', b):
      ropani = float(re.findall(r'\b(\d+)-\d+-\d+-\d+\b', b)[0])
      aana = float(re.findall(r'\b\d+-(\d+)-\d+-\d+\b', b)[0])
      paisa = float(re.findall(r'\b\d+-\d+-(\d+)-\d+\b', b)[0])
      dam = float(re.findall(r'\b\d+-\d+-\d+-(\d+)\b', b)[0])
      area = ropani * 508.737047 + aana * 31.79606544 + paisa * 7.94901636 + dam * 1.98725409
    elif re.match(r'\b\d+[-.]\d+[-.]\d+\b', b):
      aana  = float(re.findall(r'\b(\d+)[-.]\d+[-.]\d+\b', b)[0])
      paisa = float(re.findall(r'\b\d+[-.](\d+)[-.]\d+\b', b)[0])
      dam = float(re.findall(r'\b\d+[-.]\d+[-.](\d+)\b', b)[0])
      area = aana * 31.79606544 + paisa * 7.94901636 + dam * 1.98725409
    else:
      area = float(b) * 508.737047

  if df.iloc[a].area_units == 'Paisa':
    if re.match(r'\b\d+-\d+-\d+-\d+\b', b):
      ropani = float(re.findall(r'\b(\d+)-\d+-\d+-\d+\b', b)[0])
      aana = float(re.findall(r'\b\d+-(\d+)-\d+-\d+\b', b)[0])
      paisa = float(re.findall(r'\b\d+-\d+-(\d+)-\d+\b', b)[0])
      dam = float(re.findall(r'\b\d+-\d+-\d+-(\d+)\b', b)[0])
      area = ropani * 508.737047 + aana * 31.79606544 + paisa * 7.94901636 + dam * 1.98725409
    elif re.match(r'\b\d+[-.]\d+[-.]\d+\b', b):
      aana  = float(re.findall(r'\b(\d+)[-.]\d+[-.]\d+\b', b)[0])
      paisa = float(re.findall(r'\b\d+[-.](\d+)[-.]\d+\b', b)[0])
      dam = float(re.findall(r'\b\d+[-.]\d+[-.](\d+)\b', b)[0])
      area = aana * 31.79606544 + paisa * 7.94901636 + dam * 1.98725409
    else:
      area = float(b) * 7.94901636

  if df.iloc[a].area_units == 'Kattha':
    if re.match(r'\b\d+-\d+-\d+\b', b):
      bigha = float(re.findall(r'\b(\d+)-\d+-\d+\b', b)[0])
      katha = float(re.findall(r'\b\d+-(\d+)-\d+\b', b)[0])
      dhur = float(re.findall(r'\b\d+-\d+-(\d+)\b', b)[0])
      area = bigha * 6772.631616 + katha * 338.6315808 + dhur * 16.93157904
    else:
      area = float(b) * 338.6315808

  if df.iloc[a].area_units == 'Dhur':
    if re.match(r'\b\d+-\d+-\d+\b', b):
      bigha = float(re.findall(r'\b(\d+)-\d+-\d+\b', b)[0])
      katha = float(re.findall(r'\b\d+-(\d+)-\d+\b', b)[0])
      dhur = float(re.findall(r'\b\d+-\d+-(\d+)\b', b)[0])
      area = bigha * 6772.631616 + katha * 338.6315808 + dhur * 16.93157904
    else:
      area = float(b) * 16.93157904

  if df.iloc[a].area_units == 'Sq. Feet':
    if re.match(r'\b\d+-\d+-\d+-\d+\b', b):
      ropani = float(re.findall(r'\b(\d+)-\d+-\d+-\d+\b', b)[0])
      aana = float(re.findall(r'\b\d+-(\d+)-\d+-\d+\b', b)[0])
      paisa = float(re.findall(r'\b\d+-\d+-(\d+)-\d+\b', b)[0])
      dam = float(re.findall(r'\b\d+-\d+-\d+-(\d+)\b', b)[0])
      area = ropani * 508.737047 + aana * 31.79606544 + paisa * 7.94901636 + dam * 1.98725409
    elif re.match(r'\b\d+-\d+-\d+\b', b):
      bigha = float(re.findall(r'\b(\d+)-\d+-\d+\b', b)[0])
      katha = float(re.findall(r'\b\d+-(\d+)-\d+\b', b)[0])
      dhur = float(re.findall(r'\b\d+-\d+-(\d+)\b', b)[0])
      area = bigha * 6772.631616 + katha * 338.6315808 + dhur * 16.93157904
    elif re.match(r'\b\d+.*\d* x \d+.*\d*\b', b):
      x = float(re.findall(r'\b(\d+.*\d*) x \d+.*\d*\b', b)[0])
      y = float(re.findall(r'\b\d+.*\d* x (\d+.*\d*)\b', b)[0])
      area = x * y * 0.09290304
    elif re.match(r'\b\d+.*\d*\*\d+.*\d*\b', b):
      x = float(re.findall(r'\b(\d+.*\d*)\*\d+.*\d*\b', b)[0])
      y = float(re.findall(r'\b\d+.*\d*\*(\d+.*\d*)\b', b)[0])
      area = x * y * 0.09290304
    else:
      area = float(b) * 0.09290304

  if df.iloc[a].area_units == 'Haat':
    if re.match(r'\b\d+.*\d*[/|\*]\d+.*\d*\b', b):
      x = float(re.findall(r'\b(\d+.*\d*)[/|\*]\d+.*\d*\b', b)[0])
      y = float(re.findall(r'\b\d+.*\d*[/|\*](\d+.*\d*)\b', b)[0])
      area = x * y * 0.20903184
    else:
      area = float(df.iloc[a].area_covered) * 0.20903184

  if df.iloc[a].area_units == 'Sq. Meter':
    if re.match(r'\b\d+.*\d*[/|\*]\d+.*\d*\b', b):
      x = float(re.findall(r'\b(\d+.*\d*)[/|\*]\d+.*\d*\b', b)[0])
      y = float(re.findall(r'\b\d+.*\d*[/|\*](\d+.*\d*)\b', b)[0])
      area = x * y
    else:
      area = float(df.iloc[a].area_covered)
  house_area.append(area)

df['area'] = house_area
df.head()
```
